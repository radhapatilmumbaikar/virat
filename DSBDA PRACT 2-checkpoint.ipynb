```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create dataset
df = pd.DataFrame({
    'Maths': [85, 90, np.nan, 95, 100, 40, 30, 200],
    'Science': [80, 85, 88, np.nan, 92, 35, 30, 150],
    'English': [78, 82, 85, 88, np.nan, 30, 25, 120]
})

print(df)

# Fill missing values
df.fillna(df.mean(), inplace=True)

# Show boxplot
sns.boxplot(data=df)
plt.show()

# Remove outliers (simple IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.5*IQR)) | (df > (Q3 + 1.5*IQR))).any(axis=1)]

print(df)

# Log transformation
df['Maths_log'] = np.log(df['Maths'])

# Graph
sns.histplot(df['Maths'])
plt.show()

sns.histplot(df['Maths_log'])
plt.show()
```
