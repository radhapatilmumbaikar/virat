```python
# ============================================
# Movie Recommendation System using Scikit-Learn
# ============================================

# 1. Import Required Libraries
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# 2. Load Dataset
df = pd.read_csv("movies.csv")


# 3. Display First 5 Rows
print("First 5 Rows of Dataset:\n")
print(df.head())


# 4. Display Dataset Information
print("\nDataset Information:\n")
print(df.info())


# 5. Select Required Columns
movies = df[['title', 'genres']]


# 6. Check Missing Values
print("\nMissing Values:\n")
print(movies.isnull().sum())


# 7. Convert Text Data into Numerical Form
cv = CountVectorizer(stop_words='english')

vector = cv.fit_transform(movies['genres'].values.astype('U'))


# 8. Calculate Cosine Similarity
similarity = cosine_similarity(vector)


# 9. Recommendation Function
def recommend(movie_name):

    # Find movie index
    index = movies[movies['title'] == movie_name].index[0]

    # Get similarity scores
    distances = similarity[index]

    # Sort movies based on similarity
    movie_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    # Display recommendations
    print("\nRecommended Movies:\n")

    for i in movie_list:
        print(movies.iloc[i[0]].title)


# 10. Test the Recommendation System
recommend('Toy Story (1995)')
```
