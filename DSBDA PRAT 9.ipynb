```python
# 1. Import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load dataset
df = sns.load_dataset("titanic")

# 3. Box plot
sns.boxplot(x="sex", y="age", hue="survived", data=df)

# 4. Title
plt.title("Age Distribution by Gender and Survival")

# 5. Show plot
plt.show()
```
